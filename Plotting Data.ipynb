```python
import numpy as np
import matplotlib.pyplot as plt
```

# Creating Basic Plots

```python
x=[1, 2, 3, 4, 5]
y=[2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)

plt.show()
```

```python
plt.plot(x, y, marker='o')

plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.title("Example Graph") 
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()
```

# Styling Plots

```python
plt.plot(x, y, marker='o')
plt.title("Example Graph") 
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y, linestyle='--', c='r')
         
plt.show()
```

```python
plt.plot(x, y, marker='s', ms=10, mfc="g")
plt.title("Example Graph") 
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y, linestyle='--', c='r')
         
plt.show()
```

```python
plt.plot(x, y, marker='*', ms=10, mfc="m")
plt.title("Example Graph") 
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y, linestyle='--', c='r')
         
plt.show()
```

```python
plt.plot(x, y, marker='d', ms=10, mfc="c")
plt.title("Example Graph") 
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y, linestyle='--', c='r')
         
plt.show()
```
