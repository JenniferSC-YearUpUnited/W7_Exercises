```python
import numpy as np
import matplotlib.pyplot as plt
```

# Creating Basic Plots

```python
x=[1, 2, 3, 4, 5]
y=[2, 3, 5, 7, 11]
```

```python
plt.plot(x, y)

plt.show()
```

```python
plt.plot(x, y, marker='o')

plt.show()
```

```python
plt.plot(x, y, marker='o')
plt.title("Example Graph") 
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()
```

# Styling Plots

```python
plt.plot(x, y, marker='o')
plt.title("Example Graph") 
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y, linestyle='--', c='r')
         
plt.show()
```

```python
plt.plot(x, y, marker='s', ms=10, mfc="g")
plt.title("Example Graph") 
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y, linestyle='--', c='r')
         
plt.show()
```

```python
plt.plot(x, y, marker='*', ms=10, mfc="m")
plt.title("Example Graph") 
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y, linestyle='--', c='r')
         
plt.show()
```

```python
plt.plot(x, y, marker='d', ms=10, mfc="c")
plt.title("Example Graph") 
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y, linestyle='--', c='r')
         
plt.show()
```

# Subplots

```python
x=np.linspace(0, 10, 100)
y=np.sin(x)
z=np.random.randn(1000)

fig, axes = plt.subplots(1, 2, figsize=(20, 10))

axes[0].plot(x,y, color="green")
axes[0].set_title("Linear Motion of a Sample")
axes[0].set_xlabel("Sequence")
axes[0].set_ylabel("Values")
                  
axes[1].hist(z, color="skyblue", edgecolor="black")
axes[1].set_title("Random Number Distribution")
axes[1].set_xlabel("Standard Deviation")
axes[1].set_ylabel("Random Values") 
axes[1].grid(axis="y", color="black", alpha=0.80)



plt.show()             
```
